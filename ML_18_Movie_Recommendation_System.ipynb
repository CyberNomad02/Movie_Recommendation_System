# Importing Dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# Data Collection and Preprocessing

```python
data = pd.read_csv('/content/movies.csv')
data.head()
```

```python
# size of data
data.shape
```

```python
# select most relevant features
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# replacing null values with null strings
for feature in selected_features:
  data[feature] = data[feature].fillna('')
```

```python
# combine features
combined_features = data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
combined_features
```

```python
# convert text data int ofeature vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

Cosine Similarity

```python
# getting similarity score using cosine similarity
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
# create a list with movies present in dataset
list_of_all_titles = data['title'].tolist()
print(list_of_all_titles)
```

```python
# getting the movie name
movie_name = input('Enter your favourite movie name: ')
```

```python
# find close match of value given by user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# find index of movie
index_of_movie = data[data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
# getting list of similar movies
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
# sorting based onm similarity score
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print(sorted_similar_movies)
```

```python
# print the name based on index
print('Movies suggested for you: \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = data[data.index==index]['title'].values[0]
  if (i<21):
    print(i,'.',title_from_index)
    i+=1
```
